```python
# Drop rows where any OHLC value is missing
df = df.dropna(subset=['open', 'high', 'low', 'close'])

# Fill missing volume with 0 (interpreted as no trade activity)
df['volume'] = df['volume'].fillna(0)
```

```python
import pandas as pd
import pandas_ta as ta


 # =========================
def process_candles_with_ta(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    df["timestamp"] = pd.to_datetime(df["timestamp"], utc=True)
    df = df.sort_values(["symbol", "timestamp"])

    all_features = []

    for symbol in df["symbol"].unique():
        #sub_df = df[df["symbol"] == symbol].copy()
        #sub_df.set_index("timestamp", inplace=True)
        sub_df = df[df["symbol"] == symbol].copy()

        # Ensure timestamps are unique per symbol
        sub_df = sub_df.drop_duplicates(subset=["timestamp"]).sort_values("timestamp")

        sub_df.set_index("timestamp", inplace=True)

        # Force numeric conversion for OHLCV columns
        for col in ["open", "high", "low", "close", "volume"]:
            sub_df[col] = pd.to_numeric(sub_df[col], errors="coerce")


        sub_df["rsi_14"] = ta.rsi(sub_df["close"], length=14)
        sub_df["ema_21"] = ta.ema(sub_df["close"], length=21)
        sub_df["ema_50"] = ta.ema(sub_df["close"], length=50)

        macd = ta.macd(sub_df["close"], fast=12, slow=26, signal=9)
        if macd is not None:
            sub_df = sub_df.join(macd)

        bbands = ta.bbands(sub_df["close"], length=20, std=2)
        if bbands is not None:
            sub_df = sub_df.join(bbands)

        sub_df["atr_14"] = ta.atr(
            high=sub_df["high"], low=sub_df["low"], close=sub_df["close"], length=14
        )

        stoch = ta.stoch(
            high=sub_df["high"], low=sub_df["low"], close=sub_df["close"], k=14, d=3
        )
        if stoch is not None:
            sub_df = sub_df.join(stoch)

        sub_df = sub_df.dropna()
        sub_df["symbol"] = symbol
        sub_df.reset_index(inplace=True)
        all_features.append(sub_df)

    final_df = pd.concat(all_features, ignore_index=True)
    final_df = final_df.sort_values(["symbol", "timestamp"]).reset_index(drop=True)
    return final_df
final_df = process_candles_with_ta(df)
# Step 5: Save or use the dataframe
final_df.to_csv("processed_candles_with_ta.csv", index=False)
print(final_df.head())
```

```python
import pandas as pd
import pandas_ta as ta


 # remove corrupted rows
df['timestamp'] = pd.to_datetime(df['timestamp'], utc=True)
df = df.sort_values(['symbol', 'timestamp'])

# Step 2: Create a processed dataframe container
all_features = []

# Step 3: Process per symbol
symbols = df['symbol'].unique()

for symbol in symbols:
    sub_df = df[df['symbol'] == symbol].copy()
    sub_df.set_index('timestamp', inplace=True)

    # Calculate indicators
    sub_df['rsi_14'] = ta.rsi(sub_df['close'], length=14)
    sub_df['ema_21'] = ta.ema(sub_df['close'], length=21)
    sub_df['ema_50'] = ta.ema(sub_df['close'], length=50)

   # MACD: Adds 3 columns: MACD_12_26_9, MACDh_12_26_9, MACDs_12_26_9
    macd = ta.macd(sub_df['close'], fast=12, slow=26, signal=9)
    if macd is not None:
        sub_df = sub_df.join(macd)

    # Bollinger Bands (20-period)
    bbands = ta.bbands(sub_df['close'], length=20, std=2)
    if bbands is not None:
        sub_df = sub_df.join(bbands)

    # ATR (Average True Range)
    sub_df['atr_14'] = ta.atr(high=sub_df['high'], low=sub_df['low'], close=sub_df['close'], length=14)

    # Stochastic Oscillator
    stoch = ta.stoch(high=sub_df['high'], low=sub_df['low'], close=sub_df['close'], k=14, d=3)
    if stoch is not None:
        sub_df = sub_df.join(stoch)

    # Drop rows with NaN from indicator computation
    sub_df = sub_df.dropna()

    # Add symbol back
    sub_df['symbol'] = symbol

    # Reset index
    sub_df.reset_index(inplace=True)

    # Append to list
    all_features.append(sub_df)


# Step 4: Concatenate all symbols
final_df = pd.concat(all_features, ignore_index=True)

# Step 5: Save or use the dataframe
final_df.to_csv("processed_candles_with_ta.csv", index=False)
print(final_df.head())
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# === CONFIG ===
sequence_length = 24   # 24 past hours
future_steps = 3     # Predict next 3 close values
features = [
    'open', 'high', 'low', 'close', 'volume', 'rsi_14', 'ema_21', 'ema_50',
    'MACD_12_26_9', 'MACDh_12_26_9', 'MACDs_12_26_9',
    'BBL_20_2.0', 'BBM_20_2.0', 'BBU_20_2.0', 'BBB_20_2.0', 'BBP_20_2.0',
    'atr_14', 'STOCHk_14_3_3', 'STOCHd_14_3_3'
]
target = 'close'

# === ASSUME final_df IS LOADED AND HAS 'symbol', 'timestamp' ===
final_df.sort_values(['symbol', 'timestamp'], inplace=True)

# === SCALE PER SYMBOL ===
scalers = {}
scaled_data_list = []
for sym in final_df['symbol'].unique():
    df_sym = final_df[final_df['symbol'] == sym].copy()
    scaler = MinMaxScaler()
    df_sym[features] = scaler.fit_transform(df_sym[features])
    scalers[sym] = scaler
    scaled_data_list.append(df_sym)

scaled_df = pd.concat(scaled_data_list, ignore_index=True)

# === CREATE SEQUENCES FOR ALL SYMBOLS ===
X, y = [], []
for sym in scaled_df['symbol'].unique():
    df_sym = scaled_df[scaled_df['symbol'] == sym]
    values = df_sym[features].values
    for i in range(sequence_length, len(values) - future_steps + 1):
        X.append(values[i-sequence_length:i])
        y.append([values[i + j][features.index(target)] for j in range(future_steps)])

X, y = np.array(X), np.array(y)
print("X shape:", X.shape, "y shape:", y.shape)

# === TRAIN/TEST SPLIT ===
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]


# === BUILD LSTM MODEL ===
model = Sequential([
    LSTM(128, return_sequences=True, input_shape=(X.shape[1], X.shape[2]),
         dropout=0.3, recurrent_dropout=0.3, kernel_regularizer=l2(0.001)),
    LSTM(64, return_sequences=True, dropout=0.3, recurrent_dropout=0.3, kernel_regularizer=l2(0.001)),
    LSTM(32, dropout=0.3, recurrent_dropout=0.3, kernel_regularizer=l2(0.001)),
    Dense(future_steps)
])
model.compile(optimizer='adam', loss='mse')

# === EARLY STOPPING ===
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# === TRAIN ===
history = model.fit(X_train, y_train, epochs=100, batch_size=32,
                    validation_data=(X_test, y_test),
                    callbacks=[early_stopping], verbose=1)

# === PREDICT ===
y_pred = model.predict(X_test)

# === METRICS PER STEP ===
for i in range(future_steps):
    rmse = np.sqrt(mean_squared_error(y_test[:, i], y_pred[:, i]))
    r2 = r2_score(y_test[:, i], y_pred[:, i])
    print(f'Step {i+1} - RMSE: {rmse:.4f}, R²: {r2:.4f}')

# === PLOT ===
plt.figure(figsize=(14, 6))
for step in range(future_steps):
    plt.plot(y_test[:, step], label=f'True t+{step+1}')
    plt.plot(y_pred[:, step], label=f'Predicted t+{step+1}', linestyle='--')
plt.legend()
plt.title("Multi-Symbol Multi-Step LSTM Predictions (Next 3 Close Prices)")
plt.show()

# === PREDICT NEXT 3 HOURS FOR EACH SYMBOL ===
predictions = {}
for sym in final_df['symbol'].unique():
    df_sym = scaled_df[scaled_df['symbol'] == sym]
    last_seq = df_sym[features].values[-sequence_length:]
    last_seq = np.expand_dims(last_seq, axis=0)
    pred_scaled = model.predict(last_seq)[0]
    scaler = scalers[sym]
    close_min = scaler.data_min_[features.index(target)]
    close_max = scaler.data_max_[features.index(target)]
    pred_unscaled = pred_scaled * (close_max - close_min) + close_min
    predictions[sym] = {f"t+{i+1}h": float(pred_unscaled[i]) for i in range(future_steps)}

print("Next 3-hour predictions per symbol:")
for sym, preds in predictions.items():
    print(sym, preds)
```
